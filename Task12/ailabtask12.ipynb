<div class="alert alert-block alert-warning">
    4.1. <b>Manage Null Values</b>

```python

df = df.drop('RAM_Size',axis=1)
```

```python

def fillnaFloatMean(cols):
    for i in cols:
        df[i] = df[i].fillna(df[i].mean())

columns = ['Processor_Speed','Storage_Capacity']
fillnaFloatMean(columns)
```

<div class="alert alert-block alert-warning">
    4.2. <b>Manage Data Types</b>
</div>

```python
df['Processor_Speed'] = df['Storage_Capacity'].astype('int64')
```

```python

def changeFloatToInt(cols):
    for i in cols:
        df[i] = df[i].astype('int64')

columns = ['Processor_Speed','Storage_Capacity']
changeFloatToInt(columns)
```

```python
df.info()
```

```python
def convertObjToInt(cols):
    for i in cols:
        data = pd.DataFrame({i:df[i].unique()})
        data_LE = LabelEncoder()
        data_LE.fit(np.ravel(data))    # data_LE.fit(data)
        df[i] = data_LE.transform(df[i])
        
columns = ['Processor_Speed','Storage_Capacity']
convertObjToInt(columns)
```

<div class="alert alert-block alert-info">
    5. <b>Train - Test Splitting</b>
</div>

```python
import pandas as pd
df = pd.read_csv('Laptop_price.csv')
df.columns
```

```python
X = df.drop('Storage_Capacity', axis=1)   
y = df['Storage_Capacity']
```

```python
from sklearn.model_selection import train_test_split
train_X, test_X, train_y, test_y = train_test_split(X,y, test_size=0.3, shuffle=False)
```

```python
print(train_X.shape, train_y.shape)
print(test_X.shape, test_y.shape)
```

<div class="alert alert-block alert-info">
    6. <b>Applying Classifier / Model Training</b>
</div>

```python
from sklearn.svm import SVC 

model_svc = SVC()
model_svc.fit(train_X, train_y)

print(model_svc)
```

```python
model_pred_svc = model_svc.predict(test_X)
model_pred_svc
```

```python
from sklearn.metrics import accuracy_score
model_acc_svc = accuracy_score(test_y, model_pred_svc)
print(round(model_acc_svc,3)*100, "%")
```
